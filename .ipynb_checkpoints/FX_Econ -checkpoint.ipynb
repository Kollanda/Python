Why delivery date for an option is different from the delivery date for a forward when we do manual  input of expiry term? Answer: For example, if you do {EUR Curncy OVML<GO>}, and enter 276D for expiry, you will get a different  delivery date from the one you'd get in {EUR Curncy OVML FWD<GO>} if you enter 276D. Essentially this is because for vanilla there's "expiry" (but not so for the forward), and the expiry  date would be 276D ahead of spot date, then there's another T+2 from expiry date to delivery date. For forward there's no "expiry", there's delivery instead. And this will be counted from spot date, in  example above 276D after spot date.  Refer to {FIFW DRQS 80370104<GO>} for explanation at length and also to {FIFW DRQS  
80326050<GO>} for similar issue related to theta involving expiry-delivery time gap There are two types of time gap in OVML:  1. the time to expiry = Expiry Date - Price Date 2. the time to delivery = Delivery Date - Premium Date When you open a screen on a vanilla EUR for example, the 276 days correspond to  the time to expiry. If you switch the instrument type to a Forward, the 276 days will be interpreted as the time to delivery, considering there is no expiry for a Forward.  When you open an OVML screen with a Vanilla, then: 1. the Price Date is read from the screen 2. the Premium Date is computed from the Price Date (usually Price Date + 2  business days, but it depends on the currency pair) 3. the Expiry Date is computed from the time to expiry (for example 276 days)  and rolled to the next business day 4. the Delivery Date is computed from the Expiry Date (usually Expiry Date + 2  business days) If you open the screen with a Forward, then: 1. the Price Date is read from the screen 2. the Premium Date is computed as for a Vanilla 3. the Delivery Date is computed from the time to delivery (276 days) and rolled to the nearest business day.


# Variance Swaps
[simple explanation](https://www.quora.com/What-is-the-difference-between-trading-variance-and-volatility-swaps) 
The variance swap replication is accomplished using a portfolio of options with different strikes. The construction of this portfolio can be understood intuitively in the Black-Scholes model - the sensitivity of a European option to the variance of the underlying asset price depends on the asset price. This "variance vega" is largest when the underlying price is closest to the strike of the option, and is also an increasing function of the strike.

The variance vega of a portfolio of options that replicates the variance swap payoff must be independent of the underlying price. To achieve this, each option has to be weighted by the inverse of the strike squared. [Inverse of strike squared explained](https://quant.stackexchange.com/questions/32951/why-is-the-fair-strike-of-a-variance-swap-called-implied-volatility)    
$\mathbf{K}_{var} =  \frac{2}{T}\Bigg(rT-\Big(\frac{S_0}{S_T}e^{rT}-1\Big)-\log{\frac{S_*}{S_0}}+e^{rT}\int_{0}^{S_*}\frac{1}{K^2}P(K)dK + e^{rT}\int_{S_*}^{\infty}\frac{1}{K^2}C(K)dK \Bigg) \tag{29} \label{formula}$


The variance exposure of the portfolio is largely independent of the underlying asset price, as long as the price lies within the range of option strikes. That is, as the spacing between strikes is decreased and the range of strikes in the options portfolio increases, the variance exposure becomes entirely independent of the underlying stock price. The rigorous derivation shows that such a portfolio indeed replicates the payoff of a variance swap.
