#General Power Series:
a_0 + a_1*x +a_2*x^2 ... + a_n*x^n = sum from N=0 to infinity ( a_n*x^n)
that's what's known as a power series because of x to power of something   
- rule of convergence 
|x|< R (radius of convergence)
(-R < x < R)

|x| > R  divergence -> |x| = R very delicate (border line and not used here) 

|a_nX^N| -> (converges to zero) exponentially fast for |x| <R 
e.g. 1/(1-x)  for |X| <1 or |X| >1 
will be the same for any power series (one case that goes up, or it is going to 0) - not considering the border case where x = 1 (not defined here)   

1+x+x^2+x^3 + .... = 1/(1-x) -> geometric series 
exactly the opposite of the power series -> can read equalities in two directions (left or right side is formula for the other) 

- very much like decimal expansion that allows you to represent numbers like the sqrt(2) - decimal expansion gives a handle on all real numbers  
[explanation](https://math.stackexchange.com/questions/409658/can-every-real-number-be-represented-by-a-possibly-infinite-decimal)
idea of series is that they are flexible enough to represent all sorts of functions -> all functions like e^x, arcsin etc

## Taylor's Formula      
f(x)  = sum n0 o inf f^(n)(0)/n!*x^n   
f of x is in terms of derivative of x -> differentiate n times, evaluate at 0 and divide by n factorial and multiply be x to the power of n 
f^(n)(0)/n! = a_n    

last is third differentiation
